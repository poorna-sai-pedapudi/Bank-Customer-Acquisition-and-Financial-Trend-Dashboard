```python
#Estimated income
df['Estimated Income'].min()

#boxing the income, like 0to20000 in a box, 20000to 40000 in a box like that.
#converting numerical data into categorical data
```

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Mid', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins = bins, labels = labels, right = False)
```

```python
df['Income Band'].value_counts()
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
#examining the distribution of categories in categorical columns

categorical_cols = df[['BRId', 'GenderId', 'IAId', 'Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure', 'Loyalty Classification', 'Properties Owned', 'Risk Weighting', 'Income Band']].columns

for col in categorical_cols:
    print(f"Value Counts for '{col}': ")
    display(df[col].value_counts())
```

Univariate Analysis, means we'll be analysing one column at a time.(Column by column)


Bivariate Analysis, means we'll be analysing 2 columns at a time.(we're adding hue, wrt 'Gender', how it behaves).

```python
#univariate Analysis
for i, predictor in enumerate(df[['BRId', 'GenderId', 'IAId', 'Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure', 'Loyalty Classification', 'Properties Owned', 'Risk Weighting', 'Income Band']].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
    #if we remove the hue, it is univariate analysis.
    #sns.countplot(data=df, x=predictor, hue='GenderId') - this is for bivariate distribution.
```

```python
#bivariate Analysis
for i, predictor in enumerate(df[['BRId', 'GenderId', 'IAId', 'Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure', 'Loyalty Classification', 'Properties Owned', 'Risk Weighting', 'Income Band']].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='GenderId')
```

```python
#bivariate Analysis
for i, predictor in enumerate(df[['BRId', 'GenderId', 'IAId', 'Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure', 'Loyalty Classification', 'Properties Owned', 'Risk Weighting', 'Income Band']].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='Nationality')
```

```python
#Histogram of value counts for different occupation

for col in categorical_cols:
    if col == 'Occupation':
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation Count')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

# Numerical Analysis

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans','Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

#Univariate analysis
plt.figure(figsize=(15,12))
for i, col in enumerate(numerical_cols):
    plt.subplot(5,2,i+1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.show()
```

```python
df.columns
```


# Heatmaps

```python
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10,10))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt = '.2f')
plt.title("Correlation Matrix")
plt.show()
```

Banking Deposits has a strong correlation with Checking Accounts and Saving Accounts.

# Insights of EDA

1. The strongest correlation occur among the "Bank Deposits" with "Checking Accounts", "Saving Accounts" and "Foreign Currency Account" indicating that customers who manintain high balances in one account type often hold substantial amount/funds across other accounts as well.
